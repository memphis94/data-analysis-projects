```python
#–û–±—ä–µ–¥–∏–Ω—è–µ–º –≤—Å–µ –¥–∞–Ω–Ω—ã–µ –≤ –æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—É users –∏ calls
pivot_users_calls = calls.pivot_table(index=['user_id', 'month'],  aggfunc={'calls_id' : 'count', 'duration' : 'sum'})
pivot_users_calls.columns = ['n_calls','duration']
#–°–±—Ä–∞—Å—ã–≤–∞–µ–º –∏–Ω–¥–µ–∫—Å, —á—Ç–æ–±—ã –Ω–µ –ø–æ—Ç–µ—Ä—è—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ 'month'
pivot_users_calls=pivot_users_calls.reset_index()
#display(pivot_users_calls)
 
#–û–±—ä–µ–¥–∏–Ω—è–µ–º –≤—Å–µ –¥–∞–Ω–Ω—ã–µ –≤ –æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—É users –∏ messages
pivot_users_messages = messages.pivot_table(index=['user_id', 'month'], values='messages_id', aggfunc='count')
pivot_users_messages.columns = ['n_messages']
#–°–±—Ä–∞—Å—ã–≤–∞–µ–º –∏–Ω–¥–µ–∫—Å, —á—Ç–æ–±—ã –Ω–µ –ø–æ—Ç–µ—Ä—è—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ 'month'
pivot_users_messages=pivot_users_messages.reset_index()
#display(pivot_users_messages.sort_values(by='user_id'))
 
#–û–±—ä–µ–¥–∏–Ω—è–µ–º –≤—Å–µ –¥–∞–Ω–Ω—ã–µ –≤ –æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—É users –∏ internet
pivot_users_internet = internet.pivot_table(index=['user_id', 'month'], values='mb_used', aggfunc='sum')
#–°–±—Ä–∞—Å—ã–≤–∞–µ–º –∏–Ω–¥–µ–∫—Å, —á—Ç–æ–±—ã –Ω–µ –ø–æ—Ç–µ—Ä—è—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ 'month'
pivot_users_internet=pivot_users_internet.reset_index()
#display(pivot_users_internet.sort_values(by='user_id'))
#print(pivot_users_internet.isna().sum())

#–û–±—ä–µ–¥–∏–Ω—è–µ–º –≤—Å–µ –≤–º–µ—Å—Ç–µ
users_all = users.merge(pivot_users_calls, on='user_id', how='left')
users_all = users_all.merge(pivot_users_messages, on=['user_id', 'month'], how='left')
users_all = users_all.merge(pivot_users_internet, on=['user_id', 'month'], how='left')
 
display(users_all.head(1000))
```

```python
#–ü–æ–ª—É—á–∏–ª–∏ –Ω—É–∂–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∑–≤–æ–Ω–∫–æ–≤, —Å–æ–æ–±—â–µ–Ω–∏–π –∏ –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω—ã–π —Ç—Ä–∞—Ñ—Ñ–∏–∫ –ø–æ –º–µ—Å—è—Ü–∞–º
def monthlycost(row):
       
    minutes_month_cost = 0
    messages_month_cost = 0
    mb_used_month_cost = 0
    month_cost=0
   
    if row['duration']>row['minutes_included']:
        minutes_month_cost=(row['duration']-row['minutes_included'])*row['rub_per_minute']
    if row['n_messages']>row['messages_included']:
        messages_month_cost=(row['n_messages']-row['messages_included'])*row['rub_per_message']
    if row['mb_used']>row['mb_per_month_included']:
        mb_used_month_cost=(row['mb_used']-row['mb_per_month_included'])*np.ceil(row['rub_per_gb']/1024)
    
    month_cost=minutes_month_cost+messages_month_cost+mb_used_month_cost+row['rub_monthly_fee']
    return month_cost

   

        
users_all_data['monthlycost']=users_all_data.apply(monthlycost, axis=1)    
    
```
